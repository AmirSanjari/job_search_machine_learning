```python
from loadData import loadData
from initAnalysis import initAnalysis
import pandas as pd 
from matplotlib import pyplot as plt
```

```python
dataFrame = loadData()
targetPlaced, dataPlaced, targetNotPlaced, dataNotPlaced = initAnalysis(dataFrame)

employmentStatusPlaced = dataPlaced["employment_status"]
employmentStatusNotPlaced = dataNotPlaced["employment_status"]


employmentStatusPlaced = employmentStatusPlaced.fillna("Unemployed").value_counts()
employmentStatusNotPlaced = employmentStatusNotPlaced.fillna("Unemployed").value_counts()


employmentStatusPlaced.plot(kind='pie', autopct='%1.0f%%')
plt.ylabel('')
plt.title('Placed - Employment Status')
plt.show()


employmentStatusNotPlaced.plot(kind='pie', autopct='%1.0f%%')
plt.title('Not Placed - Employment Status')
plt.ylabel('')
plt.show()



combinedDataFrame = pd.DataFrame({
    'Placed': employmentStatusPlaced,
    'Not Placed': employmentStatusNotPlaced
})

combinedDataFrame.plot(kind='bar')

plt.xlabel('Employment Status')
plt.ylabel('Counts')
plt.title('Counts of Employment Status')
plt.show()

```
