```python
from loadData import loadData
from initAnalysis import initAnalysis
import pandas as pd 
from matplotlib import pyplot as plt
```

```python
dataFrame = loadData()
targetPlaced, dataPlaced, targetNotPlaced, dataNotPlaced = initAnalysis(dataFrame)

challengesPlaced = dataPlaced["biggest_challenge_in_search"]
challengesNotPlaced = dataNotPlaced["biggest_challenge_in_search"]


challengesPlaced.value_counts().plot(kind='pie', autopct='%1.0f%%')
plt.ylabel('')
plt.title('Placed - Challenges In Search')
plt.show()


challengesNotPlaced.value_counts().plot(kind='pie', autopct='%1.0f%%')
plt.title('Not Placed - Challenges In Search')
plt.ylabel('')
plt.show()



challengesPlaced = challengesPlaced.value_counts()
challengesNotPlaced = challengesNotPlaced.value_counts()


combinedDataFrame = pd.DataFrame({
    'Placed': challengesPlaced,
    'Not Placed': challengesNotPlaced
})

combinedDataFrame.plot(kind='bar')

plt.xlabel('Challenges')
plt.ylabel('Counts')
plt.title('Counts of Challenge In Search')
plt.show()
```
